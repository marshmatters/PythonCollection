```python
def handle_non_numerical_data(df1):
    columns = df1.columns.values
    
    for column in columns:
        text_digit_vals = {} 
        def convert_to_int(val):
            return text_digit_vals[val]
        if df1[column].dtype != np.int64 and df1[column].dtype != np.float64:
            column_contents = df1[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0 * 1.0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1.0
            df1[column] = list(map(convert_to_int, df1[column]))
        
    return df1
df1 = handle_non_numerical_data(df1)
```
