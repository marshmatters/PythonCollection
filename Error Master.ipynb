```python
import csv
from openpyxl import Workbook
```

```python
wb1 = Workbook()
ws1 = wb1.active

with open('DailyErrorReport 2017-01.csv', 'r') as f1:
    for row in csv.reader(f1):
        ws1.append(row)
wb1.save('Error1.xlsx')

things = ['2','3','4','5','6','7','8']

for thing in things: 
    wb = Workbook()
    ws = wb.active
    with open('DailyErrorReport 2017-0'+thing+'.csv', 'r') as f:
        for row in csv.reader(f):
            ws.append(row)
    wb.save('Error'+thing+'.xlsx')
```

```python
import pandas as pd

df1 = pd.read_excel("Error1.xlsx")
df1.insert(0, 'Month', '012017')
df2 = pd.read_excel("Error2.xlsx")
df2.insert(0, 'Month', '022017')
df3 = pd.read_excel("Error3.xlsx")
df3.insert(0, 'Month', '032017')
df4 = pd.read_excel("Error4.xlsx")
df4.insert(0, 'Month', '042017')
df5 = pd.read_excel("Error5.xlsx")
df5.insert(0, 'Month', '052017')
df6 = pd.read_excel("Error6.xlsx")
df6.insert(0, 'Month', '062017')
df7 = pd.read_excel("Error7.xlsx")
df7.insert(0, 'Month', '072017')
df8 = pd.read_excel("Error8.xlsx")
df8.insert(0, 'Month', '082017')

df1=df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
df2=df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
df3=df3.loc[:, ~df3.columns.str.contains('^Unnamed')]
df4=df4.loc[:, ~df4.columns.str.contains('^Unnamed')]
df5=df5.loc[:, ~df5.columns.str.contains('^Unnamed')]
df6=df6.loc[:, ~df6.columns.str.contains('^Unnamed')]
df7=df7.loc[:, ~df7.columns.str.contains('^Unnamed')]
df8=df8.loc[:, ~df8.columns.str.contains('^Unnamed')]
```

```python
df = pd.concat([df1,df2,df4,df5,df6,df7,df8])
```
