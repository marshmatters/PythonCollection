```python
import csv
from openpyxl import Workbook
```

```python
wb1 = Workbook()
ws1 = wb1.active


things = ['1','2','3','4','5','6','7','8']

for thing in things: 
    wb = Workbook()
    ws = wb.active
    with open('DailyItemModified.2017-0'+thing+'.csv', 'r') as f:
        for row in csv.reader(f):
            ws.append(row)
    wb.save('Modified'+thing+'.xlsx')
```
